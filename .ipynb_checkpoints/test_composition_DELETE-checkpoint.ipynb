```python
df_nativ11 = pd.read_csv("data/worldfacts/2011/nativity.csv", sep=";", converters={'Country': str.strip})
```

```python
df_nativ11 = df_nativ11.drop(columns=df_nativ11.columns[[1,3]])
```

```python
df_nativ11.Country = df_nativ11.Country.replace("Korea, South", "South Korea")
df_nativ11.Country = df_nativ11.Country.replace("Congo, Democratic Republic of the", "Congo (Kinshasa)")
df_nativ11.Country = df_nativ11.Country.replace("Congo, Republic of the", "Congo (Brazzaville)")
df_nativ11.Country = df_nativ11.Country.replace("Burma", "Myanmar")
df_nativ11.Country = df_nativ11.Country.replace("Cote d'Ivoire", "Ivory Coast")
```

```python
df_nativ11 = df_nativ11.set_index("Country", drop=True, verify_integrity=True)
```

```python
df_nativ11.head()
```

```python
df_bonheur = df_bonheur.join(df_nativ11, validate="1:1")
```

```python
df_bonheur.birth_rate.astype(float)
```
