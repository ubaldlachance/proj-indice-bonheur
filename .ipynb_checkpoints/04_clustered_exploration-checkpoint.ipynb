# Exploration des corrélations par regroupement de pays.

Ce notebook sert à évaluer le même genre de corrélations que dans le notebook "03_exploration".  
Cependant les corrélations seront évaluées par regroupement de pays.  
L'objectif est de vérifier si certains type de pays ont des corrélations différentes avec les indicateurs selon leur classification.  
Le notebook utilise en entrée les données provenant du fichier "" et produira le fichier "data/merged/global_cluster.csv" qui contiendra additionnellement les informations de classification par segment.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from math import log, sqrt
from statistics import mean

from decimal import Decimal
```

```python
df_global = pd.read_csv("data/merged/global2.csv", sep=";") 
```

## Clustering (regroupement)

```python
# On prend un data frame sans l'année et le pays
df_naif = df_global.drop(columns=["annee","Country"])
```

### Avec K = 3 regroupements

```python
# Segmentation en 3 regroupements.
kmeans3 = KMeans(n_clusters=3, random_state=84, n_init=3).fit(df_naif)
```

```python
print("Nombre d'observations par regroupement")
print(f"regroupement 0: {df_global.loc[kmeans3.labels_ ==0].shape[0]}")
print(f"regroupement 1: {df_global.loc[kmeans3.labels_ ==1].shape[0]}")
print(f"regroupement 2: {df_global.loc[kmeans3.labels_ ==2].shape[0]}")
```

L'inertie représente la somme des distances au carré entre les échantillons et le centre de leur regroupement.

```python
inertia3 = kmeans3.inertia_
inertia3
```

Le coefficient de silhouette est calculé à partir de la moyenne de la distance intra-regroupement et la moyenne de la distance avec le regroupement (autre) le plus rapproché pour chaque échantillons.

```python
print(f"Coefficient de silhouette global: {silhouette_score(df_naif, kmeans3.labels_)}")
sil_samps = silhouette_samples(df_naif, kmeans3.labels_)
print(f"Coefficient de silhouette regroupement 0: {mean(sil_samps[kmeans3.labels_ == 0])}")
print(f"Coefficient de silhouette regroupement 1: {mean(sil_samps[kmeans3.labels_ == 1])}")
print(f"Coefficient de silhouette regroupement 2: {mean(sil_samps[kmeans3.labels_ == 2])}")
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
couleurs = ["yellow","green","blue"]

inf = 10
for i in range(3):
    # Aggregate the silhouette scores for samples belonging to
    # cluster i, and sort them
    reg_i_silhouettes = sil_samps[kmeans3.labels_ == i]

    reg_i_silhouettes.sort()

    len_reg_i = reg_i_silhouettes.shape[0]
    supp = inf + len_reg_i

    ax1.fill_betweenx(
        np.arange(inf, supp),
        0,
        reg_i_silhouettes,
        facecolor=couleurs[i],
        edgecolor=couleurs[i],
        alpha=0.7,
    )

    # Label the silhouette plots with their cluster numbers at the middle
    ax1.text(-0.05, inf + 0.5 * len_reg_i, str(i))

    inf = supp + 10  # espacement de 10 positions entre les regroupements

ax1.set_title("Graphique de silhouette pour 3 regroupements.")
ax1.set_xlabel("Coefficient de silhouette")
ax1.set_ylabel("Regroupement")

# The vertical line for average silhouette score of all the values
ax1.axvline(x=silhouette_score(df_naif, kmeans3.labels_), color="red", linestyle="--")

ax1.set_yticks([])  
ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

plt.show()
```

## Normalisation des pays au sein des regroupements
Les regroupements ont été faits par année/pays. Dans cette section, nous allons normaliser cette situation afin que toutes les années pour un pays se trouvent dans un seul et unique regroupement.

### Pays étant assigné à plus d'un regroupement avec K = 3

```python
# Assignation du cluster dans le dataframe
df_global["cluster_init"] = kmeans3.labels_
```

```python
df_multi_clusters =  df_global[["Country","cluster_init"]].drop_duplicates().groupby(by=["Country"]).filter(lambda row: row.cluster_init.count() > 1).sort_values("Country")
df_multi_clusters
```

Comme nous pouvons le voir, certains pays font sont dans plusieurs clusters différents.
Nous allons donc normaliser cette situation en déplaçant toutes les entrées d'un même pays dans un même cluster

Voici un exemple pour la Bulgarie.

```python
# Note le count permet de compté le nombre de lignes par cluster, les valeurs sont ensuite trié par années de manière descendente. On prend le premier donc le cluster le plus fréquent pour le pays.
df_global[(df_global.Country=="Bulgaria")].groupby("cluster_init").count().sort_values("annee", ascending=False).index[0]
```

```python
df_global.loc[df_global["Country"]=="Bulgaria", "cluster_init"]
```

```python
# Méthode pour normaliser les regroupements
def normaliser_clustering(df_dt: pd.DataFrame, df_multi: pd.DataFrame, col_name:str) -> pd.DataFrame:
    # Nouvelle colonne pour cluster
    df_dt[col_name] = df_dt["cluster_init"]
    for pays in df_multi.Country.unique():
        df_dt.loc[df_dt["Country"]==pays, col_name] = df_dt[df_dt.Country==pays].groupby("cluster_init").count().sort_values("annee", ascending=False).index[0]

    return df_dt
```

```python
df_global = normaliser_clustering(df_global, df_multi_clusters, col_name="cluster3")
```

Il ne devrait plus y avoir de pays ayant des années sur plus d'un regroupement.  
On vérifie que c'est bien le cas.

```python
df_global[["Country","cluster3"]].drop_duplicates().groupby(by=["Country"]).filter(lambda row: row.cluster3.count() > 1).sort_values("Country")
```

```python
print("Nombre d'observations par regroupement après normalisation")
print(f"regroupement 0: {df_global.loc[df_global.cluster3 ==0].shape[0]}")
print(f"regroupement 1: {df_global.loc[df_global.cluster3 ==1].shape[0]}")
print(f"regroupement 2: {df_global.loc[df_global.cluster3 ==2].shape[0]}")
```

### Liste des pays par regroupement pour 3 regroupements

```python
print("cluster 0")
print(df_global.loc[df_global.cluster3 ==0].Country.unique())

print("cluster 1")
print(df_global.loc[df_global.cluster3 ==1].Country.unique())

print("cluster 2")
print(df_global.loc[df_global.cluster3 ==2].Country.unique())
```

En gros, on peut déceler que le regroupement 1 contient plus ou moins des centres financiers relativement peu peuplés.  
Le regroupement 0 représente en gros les pays plus industrialisés, alors que le cluster 2 contient les autres pays.

On comprend donc pourquoi l'indice de silhouette réaugmentait lorsque l'on passait de 5 à 6 regroupements.  
Les trois regroupements avec k=3 ont chacun été éclaté en 2 regroupements chacun.

## Évaluations des corrélations des variables avec l'indice de bonheur

### Natalité

#### Pour trois regroupements

```python
fig, axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True, sharex=True)
axs[0].scatter(df_global[df_global.cluster3==0].nativity, df_global[df_global.cluster3==0]["Happiness Score"], s=10)
axs[1].scatter(df_global[df_global.cluster3==1].nativity, df_global[df_global.cluster3==1]["Happiness Score"], s=10)
axs[2].scatter(df_global[df_global.cluster3==2].nativity, df_global[df_global.cluster3==2]["Happiness Score"], s=10)

axs[0].set_ylabel('Indice bonheur')
for ax in axs:
    ax.set_xlabel('Natalité /1000 pop')
for i in range(0, len(axs)):
    axs[i].set_title(f'regroupement {i}')
```

```python
for i in range(0,3):
    df_temp = df_global[(df_global.cluster3==i) & (df_global.nativity.notnull())]
    c_nat = stats.pearsonr(df_temp.nativity, df_temp["Happiness Score"])
    print(f"r2 regroupement {i}: {c_nat.statistic**2}")
```

```python
for i in range(0,6):
    df_temp = df_global[(df_global.cluster6==i) & (df_global.milit.notnull())]
    c_nat = stats.pearsonr(df_temp.milit/100 * df_temp.gdp, df_temp["Happiness Score"])
    print(f"r2 regroupement {i}: {c_nat.statistic**2}")
```

Pour les indicateurs des dépenses en santé et en éducation, on peut observer certaines corrélations, surtout pour les pays des regroupements 0 et 1.

Pour l'indicateur sur les taxes et autres revenus, il y a une certaine corrélation lorsque calculée globalement, cependant, on n'observe pas ce même phénomène lorsque l'on regarde par regroupement.

Pour la plupart des indicateurs, il y a des différences notables entre des segments qui composaient les mêmes segments lorsque l'on avait seulement trois indicateurs.
