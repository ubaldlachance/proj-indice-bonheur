# Exploration des corrélations par regroupement de pays.

Ce notebook sert à évaluter le même genre de corrélations que dans le notebook "exploration".
Cependant les corrélations seront évalué par groupe de pays.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from math import log, sqrt
from statistics import mean

from decimal import Decimal
```

```python
df_global = pd.read_csv("data/merged/global2.csv", sep=";") 
```

## Clustering (regroupement)

```python
df_naif = df_global.drop(columns=["annee","Country"])
```

### Avec K = 3 regroupements

```python
# Segmentation en 3 regroupements.
kmeans3 = KMeans(n_clusters=3, random_state=84, n_init=3).fit(df_naif)
```

```python
print("Nombre d'observations par cluster")
print(f"cluster 0: {df_global.loc[kmeans3.labels_ ==0].shape[0]}")
print(f"cluster 1: {df_global.loc[kmeans3.labels_ ==1].shape[0]}")
print(f"cluster 2: {df_global.loc[kmeans3.labels_ ==2].shape[0]}")
```

L'inertie représente la somme des distances au carré entre les échantillons et le centre de leur regroupement.

```python
inertia3 = kmeans3.inertia_
inertia3
```

Le coéfficient de silhouette est calculé à partir de la moyenne de la distance intra-regroupment et la moyenne de la distance avec le regroupement(autre) le plus raproché pour chaques échantillons.

```python
print(f"Coéfficient de silhouette global: {silhouette_score(df_naif, kmeans3.labels_)}")
sil_samps = silhouette_samples(df_naif, kmeans3.labels_)
print(f"Coéfficient de silhouette regroupement 0: {mean(sil_samps[kmeans3.labels_ == 0])}")
print(f"Coéfficient de silhouette regroupement 1: {mean(sil_samps[kmeans3.labels_ == 1])}")
print(f"Coéfficient de silhouette regroupement 2: {mean(sil_samps[kmeans3.labels_ == 2])}")
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
couleurs = ["yellow","green","blue"]

inf = 10
for i in range(3):
    # Aggregate the silhouette scores for samples belonging to
    # cluster i, and sort them
    reg_i_silhouettes = sil_samps[kmeans3.labels_ == i]

    reg_i_silhouettes.sort()

    len_reg_i = reg_i_silhouettes.shape[0]
    supp = inf + len_reg_i

    ax1.fill_betweenx(
        np.arange(inf, supp),
        0,
        reg_i_silhouettes,
        facecolor=couleurs[i],
        edgecolor=couleurs[i],
        alpha=0.7,
    )

    # Label the silhouette plots with their cluster numbers at the middle
    ax1.text(-0.05, inf + 0.5 * len_reg_i, str(i))

    inf = supp + 10  # espacement de 10 positions entre les regroupements

ax1.set_title("Graphique de silhouette pour 3 regroupements.")
ax1.set_xlabel("Coefficient de silhouette")
ax1.set_ylabel("Regroupement")

# The vertical line for average silhouette score of all the values
ax1.axvline(x=silhouette_score(df_naif, kmeans3.labels_), color="red", linestyle="--")

ax1.set_yticks([])  
ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

plt.show()
```

### Pays étant assigné à plus d'un cluster avec K = 3

```python
# Assignation du cluster dans le dataframe
df_global["cluster_init"] = kmeans3.labels_
```

```python
df_multi_clusters =  df_global[["Country","cluster_init"]].drop_duplicates().groupby(by=["Country"]).filter(lambda row: row.cluster_init.count() > 1).sort_values("Country")
df_multi_clusters
```

Comme nous pouvons le voir, certain pays font sont dans plusieurs clusters différents.
Nous allons donc normaliser cette situation en déplacant toutes les entrées d'un même pays dans un même cluster

```python
# Note le count permet de compté le nombre de lignes par cluster, les valeurs sont ensuite trié par années de manière descendente. On prend le premier donc le cluster le plus fréquent pour le pays.
df_global[(df_global.Country=="Bulgaria")].groupby("cluster_init").count().sort_values("annee", ascending=False).index[0]
```

```python
df_global.loc[df_global["Country"]=="Bulgaria", "cluster_init"]
```

```python
def normaliser_clustering(df_dt: pd.DataFrame, df_multi: pd.DataFrame, col_name:str) -> pd.DataFrame:
    # Nouvelle colonne pour cluster
    df_dt[col_name] = df_dt["cluster_init"]
    for pays in df_multi.Country.unique():
        df_dt.loc[df_dt["Country"]==pays, col_name] = df_dt[df_dt.Country==pays].groupby("cluster_init").count().sort_values("annee", ascending=False).index[0]

    return df_dt
```

```python
df_global = normaliser_clustering(df_global, df_multi_clusters, col_name="cluster3")
```

Il ne devrait plus y avoir de pays ayant des années sur plus d'un cluster.  
On vérifie que c'est bien le cas.

```python
df_global[["Country","cluster3"]].drop_duplicates().groupby(by=["Country"]).filter(lambda row: row.cluster3.count() > 1).sort_values("Country")
```

```python
print("Nombre d'observations par cluster après normalisation")
print(f"cluster 0: {df_global.loc[df_global.cluster3 ==0].shape[0]}")
print(f"cluster 1: {df_global.loc[df_global.cluster3 ==1].shape[0]}")
print(f"cluster 2: {df_global.loc[df_global.cluster3 ==2].shape[0]}")
```

## Évaluations des corrélations des variables avec l'indice de bonheur

### Natalité

#### Pour trois regroupements

```python
fig, axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True, sharex=True)
axs[0].scatter(df_global[df_global.cluster3==0].nativity, df_global[df_global.cluster3==0]["Happiness Score"], s=10)
axs[1].scatter(df_global[df_global.cluster3==1].nativity, df_global[df_global.cluster3==1]["Happiness Score"], s=10)
axs[2].scatter(df_global[df_global.cluster3==2].nativity, df_global[df_global.cluster3==2]["Happiness Score"], s=10)

axs[0].set_ylabel('Indice bonheur')
for ax in axs:
    ax.set_xlabel('Natalité /1000 pop')
for i in range(0, len(axs)):
    axs[i].set_title(f'cluster {i}')
```

```python
for i in range(0,3):
    df_temp = df_global[(df_global.cluster3==i) & (df_global.nativity.notnull())]
    c_nat = stats.pearsonr(df_temp.nativity, df_temp["Happiness Score"])
    print(f"r2 cluster {i}: {c_nat.statistic**2}")
```

```python
df_temp = df_global[(df_global.nativity.notnull())]
c_nat = stats.pearsonr(df_temp.nativity, df_temp["Happiness Score"])
print(f"r2 global: {c_nat.statistic**2}")
```

#### Avec cluster 1 et 2

```python
df_temp = df_global[((df_global.cluster3==1) | (df_global.cluster3==2)) & (df_global.nativity.notnull())]
c_nat = stats.pearsonr(df_temp.nativity, df_temp["Happiness Score"])
print(f"r2 cluster 1 et 2: {c_nat.statistic**2}")
```

#### Pour six regroupements

```python
for i in range(0,6):
    df_temp = df_global[(df_global.cluster6==i) & (df_global.nativity.notnull())]
    c_nat = stats.pearsonr(df_temp.nativity, df_temp["Happiness Score"])
    print(f"r2 cluster {i}: {c_nat.statistic**2}")
```

Corrélation très élevé pour les regroupements 1 et 3, beaucoup plus élevé que pour le regroupement 1 en k=3.

## En conclusion

### Avec trois regroupements

Pour les indicateurs des dépenses en santé et en éducation on peut observer certaines corrélations, surtout pour les pays des regroupements 0 et 1.  
Pour l'indicateur sur la natalité, la corrélation est beaucoup plus grande pour les pays du regroupement 2.  
Pour l'indicateur sur les taxes et autres revenus, il y a une certaine corrélation lorsque calculée globalement, cependant on n'observe pas ce même phénomène lorsque l'on regarde par regroupement.

```python
### Avec six regroupements
```

Pour la plupart des indicateurs, il y a des différences notables entre des segments qui composait les même segments lorsque l'on avait seulement trois indicateurs.  
Faire six regroupements semble effectivement apporter une certaine valeur ajouté pour ce qui est de classifier les pays.
