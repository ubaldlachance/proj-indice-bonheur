Ce notebook sert à faire la composition du jeu de données des différences entre la première et la dernière année disponible à partir du jeu de données des indices de bonheur et indicateurs.  
Fichier de données requis: data/merged/global_cluster.csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.cluster import KMeans
from math import log, sqrt
```

```python
df_global = pd.read_csv("data/merged/global_cluster.csv", sep=";") 
```

```python
df_global[df_global.Country=="Afghanistan"].sort_values(by="annee", ascending=False)
```

```python
def creer_df_diffs(df_in: pd.DataFrame) -> pd.DataFrame:
    lst_out = []
    df_in = df_in.rename(columns={"Happiness Score": "Indice_bonheur"})
    
    # Boucle sur tout les pays qui apparaissent plus d'une année. 
    for tup in df_global.groupby("Country").agg({"annee": ["min", "max"]}).reset_index().itertuples():
        # On s'assurre qu'il y a plus d'une année
        if tup[2] != tup[3]:
            enreg_deb = df_in.loc[(df_in.Country == tup[1]) & (df_in.annee == tup[2])].iloc[0]
            enreg_fin = df_in.loc[(df_in.Country == tup[1]) & (df_in.annee == tup[3])].iloc[0]
            
            lst_out.append({
                "Pays": tup[1],
                "Indice_bonheur": enreg_fin.Indice_bonheur - enreg_deb.Indice_bonheur,
                "Natalité": enreg_fin.nativity - enreg_deb.nativity,
                "Migration": enreg_fin.migration - enreg_deb.migration,
                "Dépenses_santé": enreg_fin.health_expend - enreg_deb.health_expend,
                "Dépenses_éduc": enreg_fin.educ_expend - enreg_deb.educ_expend,
                "PIB": enreg_fin.gdp - enreg_deb.gdp,
                "Taxes_rev": enreg_fin.taxes - enreg_deb.taxes,
                "Budget_equil": enreg_fin.budget - enreg_deb.budget,
                "Dépenses_milit": enreg_fin.milit - enreg_deb.milit,
                "Regroupement": enreg_fin.cluster
            })
            
    return pd.DataFrame(lst_out)
```

```python
df_diffs = creer_df_diffs(df_global)
```

## Nouveau clustering (regroupement)

Ici nous allons faire un nouveau regroupement basé directement sur le jeu de données des différences.

```python
df_naif = df_diffs.drop(columns=["Pays"])
```

```python
# Segmentation en 4 regroupements.
kmeans = KMeans(n_clusters=2, random_state=84, n_init=3).fit(df_naif)
```

```python
print("Nombre d'observations par cluster")
print(f"cluster 0: {df_diffs.loc[kmeans.labels_ ==0].shape[0]}")
print(f"cluster 1: {df_diffs.loc[kmeans.labels_ ==1].shape[0]}")
```

```python
# Assignation du cluster dans le dataframe
df_diffs["cluster_diffs"] = kmeans.labels_
```

### Liste des pays par clusters

```python
print("cluster 0")
print(df_diffs.loc[df_diffs.cluster_diffs ==0].Pays.unique())

print("cluster 1")
print(df_diffs.loc[df_diffs.cluster_diffs ==1].Pays.unique())
```

## Évaluations des corrélations entre les différences des variables et celles de l'indice de bonheur

### Natalité

```python
plt.scatter(df_diffs.Natalité, df_diffs["Indice_bonheur"], s=10)
plt.ylabel('Indice bonheur')
plt.xlabel('Natalité /1000 pop')
plt.show()
```

```python
df_temp = df_diffs[df_diffs.Natalité.notnull()]
c_nat = stats.pearsonr(df_temp.Natalité, df_temp["Indice_bonheur"])
c_nat.statistic**2
```

#### par regroupements (selon le 04)

```python
fig, axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True, sharex=True)
axs[0].scatter(df_diffs[df_diffs.Regroupement==0].Natalité, df_diffs[df_diffs.Regroupement==0]["Indice_bonheur"], s=10)
axs[1].scatter(df_diffs[df_diffs.Regroupement==1].Natalité, df_diffs[df_diffs.Regroupement==1]["Indice_bonheur"], s=10)
axs[2].scatter(df_diffs[df_diffs.Regroupement==2].Natalité, df_diffs[df_diffs.Regroupement==2]["Indice_bonheur"], s=10)

axs[0].set_ylabel('Indice bonheur')
for ax in axs:
    ax.set_xlabel('Natalité /1000 pop')
for i in range(0, len(axs)):
    axs[i].set_title(f'Regroupement {i}')
```

```python
for i in range(0,3):
    df_temp = df_diffs[(df_diffs.Regroupement==i) & (df_diffs.Natalité.notnull())]
    c_nat = stats.pearsonr(df_temp.Natalité, df_temp["Indice_bonheur"])
    print(f"r2 cluster {i}: {c_nat.statistic**2}")
```

```python
fig, axs = plt.subplots(1, 2, figsize=(9, 3), sharey=True, sharex=True)
axs[0].scatter(df_diffs[df_diffs.cluster_diffs==0].Natalité, df_diffs[df_diffs.cluster_diffs==0]["Indice_bonheur"], s=10)
axs[1].scatter(df_diffs[df_diffs.cluster_diffs==1].Natalité, df_diffs[df_diffs.cluster_diffs==1]["Indice_bonheur"], s=10)

axs[0].set_ylabel('Indice bonheur')
for ax in axs:
    ax.set_xlabel('Natalité /1000 pop')
for i in range(0, len(axs)):
    axs[i].set_title(f'Regroupement {i}')
```
