# Composition du jeu de données des différences

Ce notebook sert à faire la composition du jeu de données des différences depuis l'année précédente à partir du jeu de données des indices de bonheur et indicateurs.  
Fichier de données en entrée : data/merged/global_cluster.csv  

Il est important de noté que l'expérimentation effectuée dans ce notebook ne s'est pas révélée profitable pour les prochaines étapes et n'a pas d'impact sur ce qui suivra.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.cluster import KMeans
from math import log, sqrt
```

```python
df_global = pd.read_csv("data/merged/global_cluster.csv", sep=";") 
```

```python
df_global.shape[0]
```

```python
# Méthode pour créer le jeu de données des différences
def creer_df_diffs(df_in: pd.DataFrame) -> pd.DataFrame:
    lst_out = []
    df_in = df_in.rename(columns={"Happiness Score": "Indice_bonheur"})
    
    # Boucle sur tous les pays qui apparaissent plus d'une année. 
    for pays in df_in[["Country"]].value_counts().loc[lambda cpt: cpt>1].reset_index(name="compte").Country:
        enreg_suivant = None
        
        # Boucle sur les années du pays, descendente
        for enreg_courrant in df_in[df_in.Country == pays].sort_values(by="annee", ascending=False).itertuples():
            if enreg_suivant:
                lst_out.append({
                    "Annee": enreg_suivant.annee,
                    "Pays": pays,
                    "Indice_bonheur": enreg_suivant.Indice_bonheur - enreg_courrant.Indice_bonheur,
                    "Natalité": enreg_suivant.nativity - enreg_courrant.nativity,
                    "Migration": enreg_suivant.migration - enreg_courrant.migration,
                    "Dépenses_santé": enreg_suivant.health_expend - enreg_courrant.health_expend,
                    "Dépenses_éduc": enreg_suivant.educ_expend - enreg_courrant.educ_expend,
                    "PIB": enreg_suivant.gdp - enreg_courrant.gdp,
                    "Taxes_rev": enreg_suivant.taxes - enreg_courrant.taxes,
                    "Budget_equil": enreg_suivant.budget - enreg_courrant.budget,
                    "Dépenses_milit": enreg_suivant.milit - enreg_courrant.milit,
                    "Regroupement": enreg_suivant.cluster6
                })
            
            enreg_suivant = enreg_courrant
            
    return pd.DataFrame(lst_out)
```

```python
df_diffs = creer_df_diffs(df_global)
```

```python
plt.scatter(df_diffs.Natalité, df_diffs["Indice_bonheur"], s=10)
plt.ylabel('Indice bonheur')
plt.xlabel('Natalité /1000 pop')
plt.show()
```

```python
df_temp = df_diffs[df_diffs.Natalité.notnull()]
c_nat = stats.pearsonr(df_temp.Natalité, df_temp["Indice_bonheur"])
c_nat.statistic**2
```

```python
fig, axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True, sharex=True)
axs[0].scatter(df_diffs[df_diffs.Regroupement==0].Natalité, df_diffs[df_diffs.Regroupement==0]["Indice_bonheur"], s=10)
axs[1].scatter(df_diffs[df_diffs.Regroupement==1].Natalité, df_diffs[df_diffs.Regroupement==1]["Indice_bonheur"], s=10)
axs[2].scatter(df_diffs[df_diffs.Regroupement==2].Natalité, df_diffs[df_diffs.Regroupement==2]["Indice_bonheur"], s=10)

axs[0].set_ylabel('Indice bonheur')
for ax in axs:
    ax.set_xlabel('Natalité /1000 pop')
for i in range(0, len(axs)):
    axs[i].set_title(f'Regroupement {i}')
```

```python
for i in range(0,3):
    df_temp = df_diffs[(df_diffs.Regroupement==i) & (df_diffs.Natalité.notnull())]
    c_nat = stats.pearsonr(df_temp.Natalité, df_temp["Indice_bonheur"])
    print(f"r2 Regroupement {i}: {c_nat.statistic**2}")
```

```python
df_naif = df_diffs.drop(columns=["Annee","Pays"])
```
